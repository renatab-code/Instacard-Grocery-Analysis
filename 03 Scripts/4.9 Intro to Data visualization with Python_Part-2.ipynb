## 03. Create a histogram of the “order_hour_of_day” column

```python
# Histogram for 'order_hour_of_day'
hist = df['order_hour_of_day'].plot.hist(bins = 24)
```

#### The histogram shows the distribution of customer orders throughout the day. Most orders are placed between 9 AM and 4 PM, with a clear peak around 10 AM to 12 PM. Early morning hours (before 7 AM) and late evening hours (after 8 PM) show significantly lower order activity. This suggests that customers tend to shop primarily during late morning and early afternoon, possibly aligning with daily routines or availability.

## 04. Create a bar chart from the “loyalty_flag” column

```python
# Creating a bar chart
bar = df['loyalty_flag'].value_counts().plot.bar()
```

## 05. Check whether there’s a difference in expenditure (the “prices” column) depending on the hour of the day

```python
#creating a sample of the df
sampled_df = df.sample(n=100000, random_state=1)
```

```python
# group by hour of the day and calculate the mean expenditure
hourly_expenditure = sampled_df.groupby('order_hour_of_day')['prices'].mean().reset_index()
```

```python
# Creating a line chart
line_price_hod = sns.lineplot(data = sampled_df, x = 'order_hour_of_day', y = 'prices')
```

## 07. Create a scatterplot using the sns.scatterplot() function
