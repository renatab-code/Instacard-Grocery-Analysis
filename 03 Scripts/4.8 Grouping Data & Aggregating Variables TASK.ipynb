#### 06. The marketing team at Instacart wants to know whether there’s a difference between the spending habits of the three types of customers you identified. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer). What you’re trying to determine is whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.

```python
# Use the 'groupby' function to split the data into groups based on the 'loyalty_flags' column
# Aggregate the data based on prices, providing basic statistics such as mean, min, max, std and median
# The results show the average prices of the products purchased by each category of customer (loyal, regular, new)
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max', 'std', 'median']})
```

#### 07.The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:

##### If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”

##### If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# Create a new column for the average spending of a customer
ords_prods_merge['average_spend'] = ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')
```

```python
# Check the new column
ords_prods_merge.head(20)
```

#### Assign spender tag to each customer

```python
ords_prods_merge.loc[ords_prods_merge['average_spend']<10,'spender_flag']='Low_spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_spend']>= 10, 'spender_flag'] = 'High_spender'
```

```python
# Check spender flags
ords_prods_merge['spender_flag'].value_counts(dropna = False)
```

##### The results show a relatively low number of high spenders, once again, likely due to the removal of outliers

#### 08. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:

##### If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”

##### If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”

##### If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# Create a new column for customer frequency flags based on the median values from 'days_since_prior_order' column
ords_prods_merge['customer_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
# Check that the column was created successfully and properly
ords_prods_merge.head(20)
```

#### Based on the criteria provided, create a column featuring order frequency flags

```python
ords_prods_merge.loc[ords_prods_merge['customer_frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['customer_frequency'] > 10) & (ords_prods_merge['customer_frequency'] <=20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['customer_frequency'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Check value counts for new column
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```
